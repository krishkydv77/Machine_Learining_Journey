# How It Works?

Logistic Regression first calculates a value and then applies a Sigmoid Function.

The sigmoid function converts any value into a probability between 0 and 1.

#Advantages

✔ Simple

✔ Fast

✔ Easy to interpret

✔ Gives probability

✔ Works well for binary classification

#Limitations

❌ Not good for very complex patterns

❌ Sensitive to outliers

❌ Assumes a roughly linear relationship between features and log-odds

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/covid_toy - covid_toy.csv')
df.head(2)
```

```python
df['fever']=df['fever'].fillna(df['fever'].mean())
df.head(2)
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
df['cough']=lb.fit_transform(df['cough'])
df['gender']=lb.fit_transform(df['gender'])
df['city']=lb.fit_transform(df['city'])
df['has_covid']=lb.fit_transform(df['has_covid'])
df.head(2)
```

```python
x=df.drop(columns=['has_covid'])
y=df['has_covid']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
