## (1). Linear Regression Model :-

- Linear Regression:-Linear Regression is a supervised machine learning algorithm used to predict a continuous numerical value.

- Linear Regression tries to fit the best straight line through the data points.

- - The equation is:

Y=mX+b

Where:

Y = Predicted value (Target)
X = Input feature
m = Slope (how much Y changes when X changes)
b = Intercept (starting value)

      ↓

      ↓

Learn Relationship

      ↓

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler ,LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge,Lasso,ElasticNet
from sklearn.metrics import r2_score,mean_squared_error
```

```python
df=pd.read_csv('/content/insurance - insurance.csv')
df.head(3)
```

```python
lb=LabelEncoder()
df['sex']=lb.fit_transform(df['sex'])
df['smoker']=lb.fit_transform(df['smoker'])
df['region']=lb.fit_transform(df['region'])

df.head()
```

```python
x=df.drop(columns=['charges'])
y=df['charges']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
ss=StandardScaler()
x_train=ss.fit_transform(x_train)
x_test=ss.transform(x_test)
```

## Ridge:-
- -Ridge adds L2 penalty
- - it reduces cofficient value but does not make them exactly zero.

```python
ridge_model=Ridge(alpha=1.0)
ridge_model.fit(x_train,y_train)
```

```python
ridge_pred=ridge_model.predict(x_test)
```

```python
ridge_mse=mean_squared_error(y_test,ridge_pred)
ridge_r2=r2_score(y_test,ridge_pred)
print("MSE:",ridge_mse)
print("R2-Score:",ridge_r2)
print("\n Coficients:",ridge_model.coef_)
```

## Lasso :-  
- lasso adds L1 penalty
- it can make some cofficients exactly zero to performs feature selection.

```python
lasso_model=Lasso(alpha=0.1)
lasso_model.fit(x_train,y_train)
```

```python
lasso_pred= lasso_model.predict(x_test)
```

```python
lasso_mse=mean_squared_error(y_test,lasso_pred)
lasso_r2=r2_score(y_test,lasso_pred)
print("MSE:",lasso_mse)
print("R2-Score:",lasso_r2)
print("\n cofficients:",lasso_model.coef_)
```

## ElasticNet combines both:-  
- L1 regularization (Lasso)
- L2 regularization (Ridge)

```python
elastic_model=ElasticNet(alpha=0.1,l1_ratio=0.5)
elastic_model.fit(x_train,y_train)
```

```python
elastic_pred=elastic_model.predict(x_test)
```

```python
elastic_mse=mean_squared_error(y_test,elastic_pred)
elastic_r2=r2_score(y_test,elastic_pred)
print("MSE:",elastic_mse)
print("R2-Score:",elastic_r2)
print("\n cofficients :",elastic_model.coef_)
```

## cross validation

```python
from sklearn.linear_model import  LassoCV
lasso_cv=LassoCV(
    alphas=[0.001,0.01,0.1,10],
    cv=5,
    random_state=42
)
lasso_cv.fit(x_train,y_train)
print("best Alpha:",lasso_cv.alpha_)
```

```python
from sklearn.linear_model import ElasticNetCV
elastic_cv=ElasticNetCV(
    alphas=[0.001,0.01,0.1,1,10],
    l1_ratio=[0.1,0.3,0.5,0.9,1],
    cv=5,
    random_state=42
)
elastic_cv.fit(x_train,y_train)
print("best Alpha:",elastic_cv.alpha_)
print("best l1_ratio:",elastic_cv.l1_ratio_)
```
