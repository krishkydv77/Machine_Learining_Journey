- - Data= x(Input data)+y(target data)
- - x-----------> x_train,x_test
- - y---------->y_train,y_test

- - -for machine training we will pass (x_train,y_train)
- - - for machine testing (x_test) ----------y_pred

- - - ex.
   
   1+1=2

   1+2=3

   1+3=4

   1+5=6

   1+6=7


   x=(1+1,1+2................)

   y=(2,3,4,6,7)

   x_train=(1+1,1+2,1+3)

   x_test=(2,3,4)

   y_train=(1+5,1+6)

   y_test=(6,7)

from sklearn.preprocessing import StandardScaler

-- StandardScaler (from scikit-learn) is a tool that scales your data.

z=(x-μ)/σ

z=standard deviations

x = original value

μ (mean) = average

σ (standard deviation) = spread

👉 StandardScaler brings the data to a standard form,

 -- using mean=0 and std=1

👉 It improves the performance of machine learning models

👉 It is especially useful in distance-based algorithms
