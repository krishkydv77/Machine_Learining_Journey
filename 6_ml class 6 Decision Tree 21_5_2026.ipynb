#Decision Tree Model --> CART(classification and Regression tree)

- when data is numerical--> branch individual output avg. --> final output
- when data is categorical --> branch individual output (100 covid patients )--> 70 output(yes) /30 output(no)--> final output(70->yes)

- loss= -p_yes *log_yes -p_no * log_no

#Decision Tree is a supervised machine learning algorithm used for both:

- Classification (Yes/No, Pass/Fail)
- Regression (Predicting numbers)

#Tree Structure

A Decision Tree has 3 main parts:

1. Root Node

First question.


2. Branches

Possible answers.

Yes/
No

3. Leaf Node

Final prediction.

Pass/
Fail

#Advantages

✔ Easy to understand

✔ Easy to visualize

✔ Works for classification and regression

✔ No need for feature scaling

✔ Handles nonlinear relationships

#Disadvantages

❌ Can overfit easily

❌ Sensitive to small changes in data

❌ Large trees become complex

##Important Terms
#Gini Impurity

Used to find the best split.

A lower Gini value means a better split.

#Entropy

Measures disorder in data.

A lower entropy means purer groups.

#Information Gain

How much information we gain after a split.

Higher Information Gain = Better Split.

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

```

1. numerical data ---> decisiontreeregression

```python
df=pd.read_csv("/content/insurance - insurance.csv")
df.head()
```

```python
lb=LabelEncoder()
df["sex"]=lb.fit_transform(df["sex"])
df["smoker"]=lb.fit_transform(df["smoker"])
df["region"]=lb.fit_transform(df["region"])
df.head()
```

```python
x=df.drop(columns="charges")
y=df["charges"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
dt=DecisionTreeRegressor()
dt.fit(x_train,y_train)
```

```python
y_pred=dt.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
