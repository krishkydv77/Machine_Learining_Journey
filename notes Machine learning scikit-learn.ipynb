- StandardScaler is a **feature scaling technique** used in Machine Learning.
It standardizes data by converting features into **mean 0 and standard deviation 1.**

  σ = Standard deviation
  μ= Mean (average)
  x= Original value
  z=Scaled value

  - - use:-Different features have different values.

## Regularization Notes
Regularization is a technique used in machine learning to reduce overfitting by adding a penalty term to the loss function.
- We use regularization to improve model performance and prevent overfitting.
#Type of Regularization:-
1. Ridge Regression :-
Ridge Regression is a regularization technique that adds L2 penalty to reduce overfitting and improve model performance.
- We use Ridge Regression to reduce overfitting and handle multicollinearity in machine learning models.
- L2 penalty :-L2 penalty is the square of coefficients added to the loss function in Ridge Regressio.

   means:-

         -reduces coefficient values

         - but does NOT make coefficients zero

2. Lasso Regression :-  
Lasso Regression is a regularization technique that adds L1 penalty to reduce overfitting and perform feature selection.
- We use Lasso Regression to reduce overfitting and remove unnecessary features from the model.
- L1 penalty:- L1 penalty is the sum of absolute values of coefficients added to the loss function in Lasso Regression.

means:-
Lasso:

- reduces coefficient values
- can make some coefficients exactly zero

This performs:

- Feature Selection
- remove unimportant features

3. ElasticNet :-   
ElasticNet is a regularization technique that combines L1 and L2 penalties to reduce overfitting and improve model performance.
- We use ElasticNet when the dataset contains many correlated features and both feature selection and regularization are needed.
- l1_ratio:-  l1_ratio controls the balance between L1 penalty and L2 penalty in ElasticNet.

# Why We Use ElasticNet?

Sometimes:

- Ridge cannot remove features
- Lasso may remove too many features

ElasticNet solves both problems by combining:

- Ridge
- Lasso

| Feature                   | Ridge | Lasso   | ElasticNet |
| ------------------------- | ----- | ------- | ---------- |
| Penalty                   | L2    | L1      | L1 + L2    |
| Feature Selection         | No    | Yes     | Yes        |
| Handles Multicollinearity | Yes   | Limited | Yes        |
| Coefficients become zero  | No    | Yes     | Yes        |
