```python
import pandas as pd
import os

class DataP:
    def __init__(self):
        self.df = None
        self.object_columns = []

    def process_file(self, file_path, output_path="processed_data.csv"):
        # Load CSV
        if not os.path.exists(file_path):
            raise FileNotFoundError("File not found at given path.")

        self.df = pd.read_csv(file_path)
        print("Data loaded successfully!")
        # object type columns
        self.object_columns = self.df.select_dtypes(include=['object']).columns.tolist()
        print("Object type columns:", self.object_columns)
        # Label Encoding
        from sklearn.preprocessing import LabelEncoder
        le = LabelEncoder()
        for col in self.object_columns:
            self.df[col] = le.fit_transform(self.df[col].astype(str))

        print("Label Encoding applied successfully!")

        #processed data
        self.df.to_csv(output_path, index=False)
        print(f"Processed data saved to {output_path}")

        return self.df, self.object_columns


processor = DataP()
df, obj_cols = processor.process_file("/content/covid_toy - covid_toy.csv",)
```
