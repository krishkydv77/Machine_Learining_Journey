#RandomForestClassifier
 -Random Forest Classifier is a supervised machine learning algorithm used for classification problems.

 #Why Random Forest?

- Decision Tree ki problem:

Decision Tree

     ↓

Can Overfit

=>Ek single tree kabhi-kabhi galat prediction kar sakta hai.

#Random Forest:

Many Decision Trees

        ↓

Majority Voting

        ↓

Final Prediction

=> Isliye Random Forest usually Decision Tree se better perform karta hai.

#Important Parameters
   -  # n_estimators

Number of trees.

RandomForestClassifier(n_estimators=100)

Means:

100 Decision Trees

=> More trees → Better accuracy (usually) but slower training.

#random_state
random_state=42

Same result every time.

#max_depth

Tree ki maximum depth.

max_depth=5

Overfitting control karne ke liye.

#Advantages

✔ High Accuracy

✔ Less Overfitting than Decision Tree

✔ Handles large datasets

✔ Works with missing patterns and noisy data

✔ Feature importance provide karta hai

#Disadvantages

❌ Training slower than one Decision Tree

❌ More memory use karta hai

❌ Harder to interpret

| Decision Tree      | Random Forest           |
| ------------------ | ----------------------- |
| Single Tree        | Many Trees              |
| Fast               | Slower                  |
| Overfitting zyada  | Overfitting kam         |
| Less Accurate      | More Accurate           |
| Easy to Understand | Difficult to Understand |


```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("/content/Attrition - Attrition.csv")
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
df['Attrition']=lb.fit_transform(df['Attrition'])
df['BusinessTravel']=lb.fit_transform(df['BusinessTravel'])
df['Department']=lb.fit_transform(df['Department'])
df['EducationField']=lb.fit_transform(df['EducationField'])
df['Gender']=lb.fit_transform(df['Gender'])
df['JobRole']=lb.fit_transform(df['JobRole'])
df['MaritalStatus']=lb.fit_transform(df['MaritalStatus'])
df['Over18']=lb.fit_transform(df['Over18'])
df['OverTime']=lb.fit_transform(df['OverTime'])
df.head(2)
```

```python
x=df.drop(columns=['Attrition'])
y=df['Attrition']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred=rf.predict(x_test)
accuracy_score(y_test,y_pred)
```
